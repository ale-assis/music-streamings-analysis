Criando uma tabela e um dataframe para praticar  

```python
table = [
    ['a',  '1.2',  '4.2' ],
    ['b',  '70',   '0.03'],
    ['x',  '5',    '0'   ],
]

df = pd.DataFrame(table)
display(df)
```

Usando etiquetas para exibir colunas de forma separada

```python
display(df[0])
display(df[1])
display(df[2])
```

Convertendo o tipo de dados de uma coluna inteira  
O tonumeric vai funcionar apenas em séries ou em colunas da tabela  
lembre-se de fazer a reatribuição

```python
df[1] = pd.to_numeric(df[1])
display(df[1])
```

### Convertendo TODAS as colunas do dataset de uma vez com `astype()`

```python
# dicionário contendo chaves(colunas) e valores(linhas, que são listas)
data = {
    'coluna1': ['1', '2', '3'],
    'coluna2': ['4', '5', '6'],
    'coluna3': ['7', '8', '9']
}

df = pd.DataFrame(data)
display(df)

# ao converter, lembre-se de fazer a reatribuição 
df = df.astype(int)

# verificando o tipo de dados de todas as colunas 
display(df.dtypes)

# verificando o tipo de dados de uma coluna apenas
display(df['coluna1'].dtype)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# Criando um DataFrame fictício
df = pd.DataFrame({
    'tempo': pd.date_range(start='1/1/2023', periods=100),
    'valor': np.random.randn(100).cumsum(),
    'categoria': np.random.choice(['A', 'B', 'C'], 100)
})

# Gráfico de Linhas
sns.lineplot(data=df, x='tempo', y='valor')
plt.show()

# Gráfico de Barras
sns.barplot(data=df, x='categoria', y='valor')
plt.show()

# Histogramas
sns.histplot(data=df, x='valor')
plt.show()

# Box Plot
sns.boxplot(data=df, x='categoria', y='valor')
plt.show()

# Violin Plot
sns.violinplot(data=df, x='categoria', y='valor')
plt.show()

# Gráfico de Dispersão
sns.scatterplot(data=df, x='valor', y='tempo')
plt.show()

# Heatmap
correlation_matrix = df.corr()
sns.heatmap(correlation_matrix, annot=True)
plt.show()
```
